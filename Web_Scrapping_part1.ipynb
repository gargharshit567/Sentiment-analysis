```python
import requests
import bs4
import pickle
import pandas as pd
import re
import string
from sklearn.feature_extraction.text import CountVectorizer
```

```python
urls=[
'https://scrapsfromtheloft.com/2018/12/15/vir-das-losing-it-transcript/',
'https://scrapsfromtheloft.com/2021/01/15/chris-rock-total-blackout-the-tamborine-extended-cut-transcript/',
'https://scrapsfromtheloft.com/2021/06/01/bo-burnham-inside-transcript/',
'https://scrapsfromtheloft.com/2020/11/27/larry-the-cable-guy-remain-seated-transcript/',
'https://scrapsfromtheloft.com/2020/11/24/craig-ferguson-just-being-honest-transcript/',
'https://scrapsfromtheloft.com/2020/11/18/kevin-hart-zero-fks-given-2020-transcript/',
'https://scrapsfromtheloft.com/2020/11/12/sam-morril-i-got-this-2020-transcript/',
'https://scrapsfromtheloft.com/2020/11/05/bill-burr-snl-monologue-2020-transcript/',
'https://scrapsfromtheloft.com/2020/11/05/john-mulaney-snl-monologue-2020-transcript/',
'https://scrapsfromtheloft.com/2020/10/10/ronny-chieng-asian-comedian-destroys-america-transcript/'
]
```

```python
def url_to_trans(url):
 res= requests.get(url)
 data=bs4.BeautifulSoup(res.text,'html')
 text=[p.text for p in data.find(class_="elementor-element elementor-element-74af9a5b elementor-widget elementor-widget-theme-post-content").find_all('p')] 
 print(url)
 return text
```

```python
transcripts=[url_to_trans(url) for url in urls]
```

```python
comedians=['vir','chris','burnham','larry','craig','kevin','sam','bill','john','ronny']
```

```python
def compress_data(data):
    return ' '.join(data)
```

```python
for i,com in enumerate(comedians):
    with open(r'E:\extra\data_science\project\transcripts/'+com+'.txt','wb') as file:
        pickle.dump(transcripts[i],file)
```

```python
combined_text=[compress_data(t) for t in transcripts]
combined_data= dict(zip(comedians,combined_text))
```

```python
data_combined={key:[value] for key,value in combined_data.items()}
```

```python
pd.set_option('max_colwidth',150)
data_df=pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns=['transcript']
data_df
```

```python
def clean_text_1(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text= re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text=re.sub('\w*\d\w*','',text)
    return text
```

```python
clean_data=pd.DataFrame(data_df.transcript.apply(clean_text_1))
```

```python
def clean_text_2(text):
    text=re.sub('[’“”…]','',text)
    text=re.sub('\n','',text)
    return text
```

```python
clean_data=pd.DataFrame(clean_data.transcript.apply(clean_text_2))
```

```python
clean_data
```

```python
cv= CountVectorizer(stop_words='english')
data_cv=cv.fit_transform(clean_data['transcript'])
data_dtm=pd.DataFrame(data_cv.toarray(),columns=cv.get_feature_names())
data_dtm.index=clean_data.index
data_dtm
```

```python
data_dtm.to_pickle(r"E:\extra\data_science\project\dtm.pkl")
```
